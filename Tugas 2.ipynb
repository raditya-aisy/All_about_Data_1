```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

path = "/content/Tugas /data0-SZL.csv"
df = pd.read_csv(path)

print(df.dtypes)
df_noid = df.iloc[:]
df_noid #menampilkan nilai keseluruhan


#No. 3 Mencari nilai Rata-rata, Standar deviasi, Skewness, Median, Energi, Min, dan Max

rata_rata = df.mean()
# Tampilkan hasil
print("\n")
print("Rata-rata untuk setiap kolom:")
print(rata_rata)

standar_deviasi = df.std()
# Tampilkan hasil
print("\n")
print("Standar deviasi untuk setiap kolom:")
print(standar_deviasi)

skewness = df.skew()
# Tampilkan hasil
print("\n")
print("Skewness untuk setiap kolom:")
print(skewness)

median = df.median()
# Tampilkan hasil
print("\n")
print("Median untuk setiap kolom:")
print(median)

energi = df.apply(lambda x: sum(x**2) / len(x))
# Tampilkan hasil
print("\n")
print("Energi untuk setiap kolom:")
print(energi)

min = df.min()
# Tampilkan hasil
print("\n")
print("min untuk setiap kolom:")
print(min)

max = df.max()
# Tampilkan hasil
print("\n")
print("max untuk setiap kolom:")
print(max)

#No. 4 Memplot 6 kolom secara acak

# Pilih 6 kolom secara acak
kolom_pilihan = random.sample(list(df.columns), 6)
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Plot sinyal untuk setiap kolom
for i, kolom in enumerate(kolom_pilihan):
    baris = i // 3
    kolom = i % 3
    axs[baris, kolom].plot(df[kolom_pilihan[i]])
    axs[baris, kolom].set_title(kolom_pilihan[i])

plt.tight_layout()
plt.show()

# Pilih 30 data dari kolom 1-30
kolom_pilihan = df.iloc[:, :30]
marker = ['*', 'o', 'v']

# Buat scatterplot
fig, ax = plt.subplots(figsize=(10, 10))

for i in range(0, 30, 10):
    kolom = kolom_pilihan.columns[i:i+10]
    ax.scatter(kolom_pilihan[kolom].mean(), kolom_pilihan[kolom].std(), marker=marker[i//10], label=f'Kolom {i+1}-{i+10}')

# Label sumbu
ax.set_xlabel('Rata-rata')
ax.set_ylabel('Standar Deviasi')
ax.legend()

# Tampilkan plot
plt.title('Scatterplot: Rata-rata vs Standar Deviasi untuk Setiap Grup Kolom')
plt.show()

# Pilih 30 data dari kolom 1-30
kolom_pilihan = df.iloc[:, :30]
marker = ['*', 'o', 'v']

# Buat scatterplot
fig, ax = plt.subplots(figsize=(10, 10))

for i in range(0, 30, 10):
    kolom = kolom_pilihan.columns[i:i+10]
    ax.scatter(kolom_pilihan[kolom].skew(), kolom_pilihan[kolom].median(), marker=marker[i//10], label=f'Kolom {i+1}-{i+10}')

# Label sumbu
ax.set_xlabel('Skewness')
ax.set_ylabel('Median')
ax.legend()

# Tampilkan plot
plt.title('Scatterplot: Skewness vs Median untuk Setiap Grup Kolom')
plt.show()
```
